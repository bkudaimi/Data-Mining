```python
#DSC 550 Assignment 3
#Bilal Kudaimi 
#April 1, 2021

#Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections
from mpl_toolkits.mplot3d import axes3d
'''
Importing and slicing the data
This analysis scenario is of a data frame of vehicles
Each row is one vehicle and each column represents one of the following vehicle variables:
Vehicle make, model, chassis shape, engine size, transmissions, fuel type, fuel consumption, and CO2 emissions
'''
cars = pd.read_csv('CO2 Emissions_Canada.csv')
make = cars['Make']
mpg_city = cars['Fuel Consumption City (L/100 km)']
mpg_hwy = cars['Fuel Consumption Hwy (L/100 km)']  
mpg_comb = cars['Fuel Consumption Comb (L/100 km)'] 
cyl = cars['Cylinders']
size = cars['Engine Size(L)']
emissions = cars['CO2 Emissions(g/km)']
```

```python
#Displaying the same analysis (of the same data) using 3 different charts
#Each chart uses appropriate, complete, and professional labels

#First chart: Scatter plot of CO2 emissions versus combined fuel consumption
#Three lines can be seen, indicating some third variable is affecting the plot 
plt.scatter(mpg_comb, emissions)
plt.xlabel('Combined Fuel Consumption (L/100km)')
plt.ylabel('CO2 emissions (g/km)')
plt.title('How Fuel Consumption Affects CO2 Emissions')
plt.show()
```

```python
#Second chart: Bar chart of CO2 emissions for each combined fuel consumption
#Not much can be seen with this chart

plt.bar(mpg_comb, emissions)
plt.xlabel('Combined Fuel Consumption (L/100km)')
plt.ylabel('CO2 Emissions (g/km)')
plt.title('How Fuel Consumption Affects CO2 Emissions')
plt.show()
```

```python
#Third chart: 3D scatterplot of fuel consumption, engine size, and CO2 emissions
#THis chart reveals how two variables affect a third one

fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.scatter(mpg_comb, size, emissions, c = 'r', marker = 'o')

ax.set_xlabel('Combined Fuel Consumption (L/100km)')
ax.set_ylabel('Engine Size (L)')
ax.set_zlabel('CO2 Emissions (g/km)')
plt.show()
```
