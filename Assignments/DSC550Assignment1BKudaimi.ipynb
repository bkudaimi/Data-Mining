```python
#1: Now that the program is known to work, an array of size [10,50] is made and the random function 
#   is used to fill it with random integers between 1 and 500 before loading it into a data frame
rand_matrix = np.random.randint(1, 501, size=(10,50))
rand_matrix_df = pd.DataFrame(rand_matrix)
rand_matrix_df
```

```python
#2: Calculating the sum of each row, sum of each column, and sum of all entries.
row_sums = np.sum(rand_matrix_df, axis = 1) 
col_sums = np.sum(rand_matrix_df, axis = 0)
all_sums = np.sum(rand_matrix_df).sum()
print('Sums of each row: {}\n'.format(row_sums))
print('Sums of each column: {}\n'.format(col_sums))
print('Sums of all entries: {}'.format(all_sums))
```

```python
#3: Determining the minimum, maximum, average, and standard deviation of all entries

minimum = np.min(rand_matrix_df).min()
maximum = np.max(rand_matrix_df).max()
average = np.mean(rand_matrix_df).mean()
standard_deviation = np.std(rand_matrix_df).std()
print('Minimum of all entries: {}\n'.format(minimum))
print('Maximum of all entries: {}\n'.format(maximum))
print('Average of all entries: {}\n'.format(average))
print('Standard deviation of all entries: {}'.format(standard_deviation))
```
